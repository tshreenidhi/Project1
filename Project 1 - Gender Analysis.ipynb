```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json
import requests
```

```python
#Load Gender files
arrest_gender_2020 = "Resources/2020_arrestee_gender.csv"
arrest_gender_2021 = "Resources/2021_arrestee_gender.csv"
arrest_gender_2022 = "Resources/2022_arrestee_gender.csv"

gender_2020_df = pd.read_csv(arrest_gender_2020)
gender_2021_df = pd.read_csv(arrest_gender_2021)
gender_2022_df = pd.read_csv(arrest_gender_2022)
```

```python
gender_2020_df
```

```python
gender_2020_df = gender_2020_df[gender_2020_df['Arrest Offense Category'] != 'Total']
gender_2020_df = gender_2020_df[gender_2020_df['Total Arrestee'].notnull()]
```

```python
#Checking Data Type
print (gender_2020_df.dtypes)
# Convert numeric columns to integers (remove commas and convert)
gender_2020_df['Total Arrestee'] = gender_2020_df['Total Arrestee'].astype(str).str.replace(',', '').astype(int)
gender_2020_df['Male'] = gender_2020_df['Male'].astype(str).str.replace(',', '').astype(int)
gender_2020_df['Female'] = gender_2020_df['Female'].astype(str).str.replace(',', '').astype(int)
```

```python
# Create bar chart
multi_plot = gender_2020_df.plot(kind="bar", figsize=(25,10))
# PandasPlot.set_xticklabels() can be used to set the tick labels as well
multi_plot.set_xticklabels(gender_2020_df["Arrest Offense Category"], rotation=90)
plt.xlabel('Crime Type')
plt.ylabel('Cases')
plt.tight_layout()
plt.ylim(0, 2000000)
plt.show()
```

```python
#Remove Total
gender_2021_df = gender_2021_df[gender_2021_df['Arrest Offense Category'] != 'Total']
gender_2021_df = gender_2021_df[gender_2021_df['Total Arrestee'].notnull()]
```

```python
#Checking Data Type
print (gender_2021_df.dtypes)
# Convert numeric columns to integers (remove commas and convert)
gender_2021_df['Total Arrestee'] = gender_2021_df['Total Arrestee'].astype(str).str.replace(',', '').astype(int)
gender_2021_df['Male'] = gender_2021_df['Male'].astype(str).str.replace(',', '').astype(int)
gender_2021_df['Female'] = gender_2021_df['Female'].astype(str).str.replace(',', '').astype(int)
```
